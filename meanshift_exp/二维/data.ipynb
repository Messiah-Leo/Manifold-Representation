```python
import numpy as np
import pandas as pd
import xlwt
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib  inline
```

```python
data1 = np.random.rand(300,2)* 100
data2 = np.random.randn(200,2)* 10 + 20
data3 = np.random.randn(200,2)* 10 + 70

data4 = np.random.rand(300,2)* 100 + 100
data5 = np.random.randn(200,2)* 5 + (120,170)
data6 = np.random.randn(200,2)* 15 + (170,120)

data7 = np.random.rand(300,2)* 100 + (200,0)
data8 = np.random.randn(200,2)* 7 + (250,50)
```

```python
data = np.concatenate([data1,data2,data3,data4,data5,data6,data7,data8])
plt.scatter(data[:,0],data[:,1], s=1)
plt.axis('off')
plt.savefig('original.png', dpi=1000, bbox_inches='tight', pad_inches=0)
```

```python
book = xlwt.Workbook(encoding='utf-8', style_compression=0)
sheet = book.add_sheet('二维', cell_overwrite_ok=True)
for count, pixel in zip(range(data.shape[0]), data):
    sheet.write(count, 0, pixel[0])
    sheet.write(count, 1, pixel[1])
book.save('data1024.xls')
```
