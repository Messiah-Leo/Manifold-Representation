```python
from sklearn.cluster import MeanShift
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import cv2
```

```python
img = cv2.imread('lena_gray.png',0)
ms = MeanShift(bandwidth=30)
ms.fit(img)
labels = ms.labels_
len(labels)
```

```python
# 将灰度图像转换为彩色图像
colored_labels = cv2.applyColorMap(labels, cv2.COLORMAP_JET)
```

```python
# 显示原始灰度图像
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# 显示聚类结果
plt.subplot(122)
plt.imshow(colored_labels)
plt.title('Clustered Image')
plt.axis('off')

# 显示图像
plt.show()
```
