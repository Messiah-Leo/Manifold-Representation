```python
import numpy as np
import pandas as pd
import xlwt
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib widget
```

```python
data1 = np.random.rand(300,3)* 100
data2 = np.random.randn(200,3)* 10 + 20
data3 = np.random.randn(200,3)* 10 + 70

data4 = np.random.rand(300,3)* 100 + 100
# data5 = np.random.randn(200,3)* 5 + (120,170)
# data6 = np.random.randn(200,3)* 15 + (170,120)

# data7 = np.random.rand(300,3)* 100 + (200,0)
# data8 = np.random.randn(200,3)* 7 + (250,50)
```

```python
data = np.concatenate([data1,data2,data3,data4])
fig = plt.figure()
# 
ax = fig.add_subplot(projection='3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.scatter3D(data[:,0], data[:,1], data[:,2], s=3)

# plt.axis('off')
plt.show()
plt.savefig('original.png', dpi=1000, bbox_inches='tight', pad_inches=0)
```

```python
book = xlwt.Workbook(encoding='utf-8', style_compression=0)
sheet = book.add_sheet('', cell_overwrite_ok=True)
for count, pixel in zip(range(data.shape[0]), data):
    sheet.write(count, 0, pixel[0])
    sheet.write(count, 1, pixel[1])
    sheet.write(count, 2, pixel[2])
book.save('data1026.xls')
```
