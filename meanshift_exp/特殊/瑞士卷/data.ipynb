```python
from sklearn.datasets import make_swiss_roll
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib widget

X, t = make_swiss_roll(n_samples=2000, noise=0.1)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, cmap=plt.cm.Spectral,edgecolors='black')
plt.show()
```

```python
x = np.delete(X, 1, axis=1)
x
```

```python
fig = plt.figure()
plt.scatter(x=x[:,0],y=x[:,1],s=1)
plt.show()
```

```python
import xlwt
book = xlwt.Workbook(encoding='utf-8', style_compression=0)
sheet = book.add_sheet('swiss', cell_overwrite_ok=True)
for count, pixel in zip(range(x.shape[0]), x):
    sheet.write(count, 0, pixel[0])
    sheet.write(count, 1, pixel[1])
book.save('data.xls')
```
