```python
import pandas as pd
import matplotlib.pyplot as plt

data1 = pd.read_csv('data1030.csv', names=['x','y'])
```

```python
from sklearn import cluster

bd = 2
ms1 = cluster.MeanShift(bandwidth=bd)
ms1.fit(data1)
labels1 = pd.DataFrame(ms1.labels_,columns=['Cluster ID'])
result1 = pd.concat((data1,labels1), axis=1)

result1.plot.scatter(x='x',y='y',c='Cluster ID',colormap='jet')
```

```python
data_c1 = ms1.cluster_centers_
# plt.scatter(x=data_c1[:,0],y=data_c1[:,1])
fig, ax = plt.subplots()
result1.plot.scatter(x='x',y='y',c='Cluster ID',colormap='jet',ax=ax)
for x,y in data_c1:
    circle = plt.Circle((x, y), bd, color='gray', fill=False)
    ax.add_artist(circle)

plt.show()
```
