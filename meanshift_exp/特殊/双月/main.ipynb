```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")

data1 = pd.read_csv('2d_data.txt', delimiter=' ', names=['x','y'])
```

```python
data1.plot.scatter(x='x',y='y')
# plt.axis('off')
```

```python
from sklearn import cluster

bd = 0.0
ms1 = cluster.MeanShift(bandwidth=bd)
ms1.fit(data1)
labels1 = pd.DataFrame(ms1.labels_,columns=['Cluster ID'])
result1 = pd.concat((data1,labels1), axis=1)

result1.plot.scatter(x='x',y='y',c='Cluster ID',colormap='jet',s=10)
# plt.axis('off')
```

```python
data_c1 = ms1.cluster_centers_
# plt.scatter(x=data_c1[:,0],y=data_c1[:,1])
fig, ax = plt.subplots()
result1.plot.scatter(x='x',y='y',c='Cluster ID',colormap='jet',ax=ax)
for x,y in data_c1:
    circle = plt.Circle((x, y), 0.03, color='gray', fill=False)
    ax.add_artist(circle)

plt.show()
```

```python
ms2 = cluster.MeanShift(bandwidth=0.11)
ms2.fit(data_c1)
plt.scatter(x=data_c1[:,0],y=data_c1[:,1],c=ms2.labels_)
```

```python
data_c2 = ms2.cluster_centers_
plt.scatter(x=data_c2[:,0],y=data_c2[:,1])
```

```python
fig, ax = plt.subplots()
result1.plot.scatter(x='x',y='y',c='Cluster ID',colormap='jet',ax=ax)
for x,y in data_c2:
    circle = plt.Circle((x, y), 0.09, color='gray', fill=False)
    ax.add_artist(circle)

plt.show()
```

```python
ms3 = cluster.MeanShift(bandwidth=0.14)
ms3.fit(data_c2)
plt.scatter(x=data_c2[:,0],y=data_c2[:,1],c=ms3.labels_)
```

```python
data_c3 = ms3.cluster_centers_
plt.scatter(x=data_c3[:,0],y=data_c3[:,1])
```
