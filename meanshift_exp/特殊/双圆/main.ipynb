```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('elliptical.txt', delimiter=' ', names=['x','y'])
```

```python
from sklearn import cluster

bd = 0.1
ms = cluster.MeanShift(bandwidth=bd)
ms.fit(data)
labels1 = pd.DataFrame(ms.labels_,columns=['Cluster ID'])
result1 = pd.concat((data,labels1), axis=1)

result1.plot.scatter(x='x',y='y',c='Cluster ID',colormap='jet')
```

```python
data_c1 = ms.cluster_centers_
# plt.scatter(x=data_c1[:,0],y=data_c1[:,1])
fig, ax = plt.subplots()
result1.plot.scatter(x='x',y='y',c='Cluster ID',colormap='jet',ax=ax)
for x,y in data_c1:
    circle = plt.Circle((x, y), bd/2, color='gray', fill=False)
    ax.add_artist(circle)

plt.show()
```
