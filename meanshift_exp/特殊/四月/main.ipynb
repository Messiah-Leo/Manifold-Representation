```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data1 = pd.read_csv('2d_data.txt', delimiter=' ', names=['x','y'])
```

```python
data2 = data1 +0.5
data = pd.concat([data1,data2])
data.reset_index(drop=True, inplace=True)
```

```python
plt.scatter(data['x'],data['y'])
plt.axis('off')
```

```python
from sklearn import cluster

bd = 0.04
ms1 = cluster.MeanShift(bandwidth=bd)
ms1.fit(data)
labels = pd.DataFrame(ms1.labels_,columns=['Cluster ID'])
result = pd.concat((data,labels), axis=1)

result.plot.scatter(x='x',y='y',c='Cluster ID',colormap='jet')
```

```python
data_c1 = ms1.cluster_centers_
# plt.scatter(x=data_c1[:,0],y=data_c1[:,1])
fig, ax = plt.subplots()
result1.plot.scatter(x='x',y='y',c='Cluster ID',colormap='jet',ax=ax)
for x,y in data_c1:
    circle = plt.Circle((x, y), bd*3/2, color='gray', fill=False)
    ax.add_artist(circle)

plt.show()
```

```python
ms2 = cluster.MeanShift(bandwidth=0.11)
ms2.fit(data_c1)
plt.scatter(x=data_c1[:,0],y=data_c1[:,1],c=ms2.labels_)
```

```python
data_c2 = ms2.cluster_centers_
plt.scatter(x=data_c2[:,0],y=data_c2[:,1])
```

```python
ms3 = cluster.MeanShift(bandwidth=0.14)
ms3.fit(data_c2)
plt.scatter(x=data_c2[:,0],y=data_c2[:,1],c=ms3.labels_)
```
