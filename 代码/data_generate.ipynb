```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import numpy as np
from scipy.stats import gaussian_kde

# make_moons
X, y = make_moons(n_samples=2000, noise=0.1, random_state=0)
X_y1 = X[y == 1]
# 
xy = np.vstack([X_y1[:, 0], X_y1[:, 1]])
z = gaussian_kde(xy)(xy)

# 
grid_x, grid_y = np.mgrid[min(X_y1[:, 0])-1:max(X_y1[:, 0])+1:100j, min(X_y1[:, 1])-1:max(X_y1[:, 1])+1:100j]
grid_z = gaussian_kde(xy)(np.vstack([grid_x.flatten(), grid_y.flatten()]))

# contour
plt.contour(grid_x, grid_y, grid_z.reshape(grid_x.shape), levels=10, cmap='YlOrRd')

# scatter
plt.scatter(X_y1[:, 0], X_y1[:, 1], c='blue', s=40, marker='+')


#  x  y 
plt.xlim(min(X_y1[:, 0])-1, max(X_y1[:, 0])+1)
plt.ylim(min(X_y1[:, 1])-0.5, max(X_y1[:, 1])+0.5)
plt.show()
```

```python
mu = np.array([1, 0.5])  # 
sigma = np.array([0.2, 0.2])  # 
num_samples = 1000  # 

# 
samples1 = np.random.normal(mu, sigma, (num_samples, 2))

# 
plt.scatter(samples1[:, 0], samples1[:, 1], s=30)
plt.scatter(X_y1[:, 0], X_y1[:, 1], s=30)
plt.show()
```

```python
mu = np.array([1, -0.4])  # 
sigma = np.array([0.15, 0.15])  # 
num_samples = 250  # 

# 
samples2 = np.random.normal(mu, sigma, (num_samples, 2))

# 
plt.scatter(samples1[:, 0], samples1[:, 1], s=30)
plt.scatter(samples2[:, 0], samples2[:, 1], s=30)
plt.scatter(X_y1[:, 0], X_y1[:, 1], s=30)
plt.show()
```

```python
combined = np.vstack((X_y1, samples1, samples2))
plt.scatter(combined[:, 0], combined[:, 1], c='blue', s=40, marker='+')
```

```python
import pandas as pd
import numpy as np

df_combined = pd.DataFrame(combined)

# Excel
df_combined.to_excel('combined_data.xlsx', index=False)
```
